```python
import pandas as pd
import requests
from sqlalchemy import create_engine
```

```python
resoources="Resources/Public_Schools.csv"
data=pd.read_csv(resoources)
data.head()
```

```python
data=data[["NCESID","NAME","ADDRESS","CITY","STATE","POPULATION","LATITUDE","LONGITUDE","NAICS_DESC","ENROLLMENT"]].copy()
data
```

```python
#Check any NA values
data.dropna(how='any')
```

```python
data.loc[data['POPULATION']==0]
```

```python
#Delete the Population which is Zero
clear_data=data.loc[data["POPULATION"]>0]
# RENAME THE COLUMNS
rename_data=clear_data.rename(columns={"NCESID":"id","NAME":"name","ADDRESS":"address","CITY":"city","STATE":"state","POPULATION":"population","LATITUDE":"latitude","LONGITUDE":"longitude","NAICS_DESC":"school_type","ENROLLMENT":"enrollment"})
```

```python
rename_data.set_index("id", inplace=True)
```

```python
rename_data
```

# Create Database Connection

```python
connection_string = "postgres:postgres@localhost:5432/public_school"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
rename_data.to_sql('public_school', con=engine, if_exists='append', index=True)
```
